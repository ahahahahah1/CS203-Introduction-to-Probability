First we will define our sample space with its respective probabilities.

```python
X = [100, 200, 300, 400, 500]
probabilities = [0.15, 0.10, 0.25, 0.35, 0.15]
print(sum(probabilities))
```

Thus our probability distribution function is valid. Now, we will randomly draw n samples.

```python
import random
#We start with 10 samples
samples = random.choices(X, probabilities, k = 10)
print(samples)
print(f'Sample mean = {sum(samples)/len(samples)}')
```

We are getting a sample. Now, I will repeat the process `m=1000` times and vary the number of samples I draw from 10 to 100000, each time multiplying it by 10. I will plot the results as a histogram to see the shape of the plot

```python
import matplotlib.pyplot as plt

m = 1000
n = [10, 100, 1000, 10000, 100000]

for j in (n):
    sample_means = []
    for i in range(m):
        samples = random.choices(X, probabilities, k = j)
        sample_means.append(sum(samples)/j)
    # print(sample_means)
    print("Sample mean = ", sum(sample_means)/m)
    #Plot the sample means)
    plt.hist(sample_means, bins = 100)
    plt.show()
```

As it can be seen, the shape of the curve approaches a bell curve as n increases. I changed the probability distribution function yet the result remained the same. Also, it can be noticed that the peak is around `325`. When we find the expectation of the given random variable,

```python
import numpy as np
exp = np.dot(X, probabilities)
print(exp)
```

Thus, we get more or less a bell shaped curve which has its peak near the expectation value of the random variable. Also, the mean of all the sample means is close to the expectation which shows that the distribution is almost a normal distribution.
